```python

# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set plotting style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (9, 6)

# %%
# Load dataset (seaborn provides iris)
df = sns.load_dataset('iris')
# Alternative: from sklearn.datasets import load_iris

# %%
# Quick inspection
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
display(df.head())

# %%
# Info and descriptive statistics
print("Info:")
display(df.info())
print("\nDescribe (numeric):")
display(df.describe())

# %%
# Pairwise scatter plot: relationships between features, colored by species
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()

# %%
# Histograms: distribution of each numeric feature
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
df[numeric_cols].hist(bins=15, layout=(2,2), edgecolor='black')
plt.suptitle("Histograms of numeric features")
plt.tight_layout()
plt.show()

# %%
# Boxplots to show distribution and outliers per species
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f'Boxplot of {col} by species')
    plt.show()

# %%
# Correlation matrix heatmap
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation matrix for iris numeric features")
plt.show()

# %%
# Simple insight output
print("Simple Insights:")
print("- Sepal length and sepal width show moderate correlation with petal length/width.")
print("- Petal length and petal width are strongly correlated (good predictors for species).")
print("- Setosa species is clearly separable using petal measurements.")
```
